## CDF 

```python
#Plot CDF of No. of Languages

counts, bin_edges = np.histogram(a_Python['No. of Languages'], bins = 10, density = True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)

#compute CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)

plt.show()
```
